### PyMuPDF Integration with LlamaIndex

Este parseo de pdfs es útil para los formularios donde existan "cajetines" para rellenar y que con el parseo default de llama index, estos textos no se extraen. A diferencia de `pymupdf4llm` para convertir el pdf a markdown, esto evita creear tablas donde no las hay.

```python
from llama_index.core import SimpleDirectoryReader
from llama_index.readers.file import PyMuPDFReader

# pyMuPDFReader instance
parser = PyMuPDFReader()

# Associate the file extension with the parser
file_extractor = {".pdf": parser}

documents = SimpleDirectoryReader(
    "../Example-Projects/Grant-Assistant/data/Solicitudes",
    file_extractor=file_extractor
).load_data()

print("Total number of documents:", len(documents))
documents
```

## Mistral OCR: Extracting a file as a Markdown


```python
from mistralai import Mistral
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
client = Mistral(api_key=os.environ["MISTRAL_API_KEY"])

def upload_pdf(filename):
  uploaded_pdf = client.files.upload(
  file={
      "file_name": filename,
      "content": open(filename, "rb"),
  },
  purpose="ocr"
  )

  # Get the URL of the uploaded file by using the id of the uploaded file
  signed_url = client.files.get_signed_url(file_id=uploaded_pdf.id)
  return signed_url.url

ocr_response_cv = client.ocr.process(
  model="mistral-ocr-latest",
  document={
    "type": "document_url",
    "document_url": upload_pdf("../Example-Projects/Streamlit-RAG-Chat/models/Anexo III_BLANCO (solicitud Exp).pdf"),
  },
  include_image_base64=True,
)
ocr_response_est = client.ocr.process(
  model="mistral-ocr-latest",
  document={
    "type": "document_url",
    "document_url": upload_pdf("../Example-Projects/Streamlit-RAG-Chat/models/Anexo-III-2023.pdf"),
  },
  include_image_base64=True,
)
```

```python
import datauri
def save_image(image):
    parsed = datauri.parse(image.image_base64)
    with open(image.id, "wb") as file:
      file.write(parsed.data)

def create_markdown_file(ocr_response, output_filename = "output.md"):
  with open(output_filename, "wt") as f:
    for page in ocr_response.pages:
      f.write(page.markdown)
      for image in page.images:
        save_image(image)

create_markdown_file(ocr_response_cv, "output_cv.md")
create_markdown_file(ocr_response_est, "output_est.md")
```

```python
def extract_text_from_pdf(file_path):
    with open(file_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        return " ".join([page.extract_text() for page in reader.pages])

bases = extract_text_from_pdf("../Orden 8-2019 bases IRPF.pdf")
resolucion = extract_text_from_pdf("../RESOLUCIÓN CAST 169_firmado.pdf")
convocatoria = extract_text_from_pdf("../Convocatoria 2023_6290.pdf")
```

```python
prompt = """
Teniendo en cuenta la siguiente convocatoria de subvención:
{convocatoria}

También ten en cuenta las siguientes bases de subvención:
{bases}

Y la resolución de la subvención:
{resolucion}

Por favor, hazme un esquema muy escueta de 4 puntos concisos

"""
```

```python
from google import genai
from dotenv import load_dotenv
import os

load_dotenv()

client = genai.Client(api_key=os.environ["GEMINI_API_KEY"])

total_tokens = client.models.count_tokens(
    model="gemini-2.0-flash",
    contents=prompt.format(
        convocatoria=convocatoria,
        bases=bases,
        resolucion=resolucion
    ),
)

print("Total de tokens:", total_tokens)
```
